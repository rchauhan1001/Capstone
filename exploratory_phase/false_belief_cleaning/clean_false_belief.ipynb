```python
import json

# Fields that contain English content
english_fields = [
    "能力\nABILITY",
    "序号\nINDEX", 
    "STORY",
    "QUESTION",
    "OPTION-A",
    "OPTION-B",
    "OPTION-C",
    "OPTION-D",
    "答案\nANSWER"
]

# Mapping to clean English field names
field_mapping = {
    "能力\nABILITY": "ABILITY",
    "序号\nINDEX": "INDEX",
    "STORY": "STORY",
    "QUESTION": "QUESTION",
    "OPTION-A": "OPTION-A",
    "OPTION-B": "OPTION-B",
    "OPTION-C": "OPTION-C",
    "OPTION-D": "OPTION-D",
    "答案\nANSWER": "ANSWER"
}

def extract_english_fields(input_file, output_file):
    with open(input_file, 'r', encoding='utf-8') as infile, \
         open(output_file, 'w', encoding='utf-8') as outfile:
        
        for line in infile:
            # Parse the JSON line
            data = json.loads(line.strip())
            
            # Extract English fields
            english_data = {}
            for original_key, clean_key in field_mapping.items():
                if original_key in data:
                    english_data[clean_key] = data[original_key]
            
            # Write to output file
            outfile.write(json.dumps(english_data, ensure_ascii=False) + '\n')

```

```python
# Usage
input_file = "False Belief Task.jsonl"
extract_english_fields(input_file, 'output.jsonl')
```
